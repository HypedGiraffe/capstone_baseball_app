```python
from pybaseball import pitching_stats
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
seasonal_pitching = pitching_stats(1980,2020, ind=1)
career_pitching = pitching_stats(1980, 2020, qual=20, ind=0)
```

```python
seasonal_pitching.head()
```

```python
for col in seasonal_pitching:
    seasonal_pitching = seasonal_pitching.dropna(thresh=4000, axis=1)
```

```python
pct_null = seasonal_pitching.isnull().sum() / len(seasonal_pitching)
missing_features = pct_null[pct_null > 0.80].index
seasonal_pitching.drop(missing_features, axis=1, inplace=True)
```

```python
career_pitching.head()
```

```python
x = career_pitching.columns
for i in range(0, len(x)):
    print(x[i])
```

```python
#Making Stat Specific Dataframes
df_career = pd.DataFrame()
df_career['id'] = career_pitching['IDfg']
df_career['g'] = career_pitching['G']
df_career['war'] = career_pitching['WAR']
df_career['era'] = career_pitching['ERA']
df_career['k/9'] = career_pitching['K/9']
```
