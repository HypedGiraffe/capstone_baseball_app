### File Used To Obtain Season-by-Season and Career Aggregate Data For Baseball Batters

```python
from pybaseball import batting_stats
import pandas as pd
import numpy as np
```

```python
#Brings in both seasonal and career batting statistics for players in the MLB from 2007 until 2022
career_batting = batting_stats(2007, 2022, qual=10, ind=0)
seasonal_batting = batting_stats(2007, 2022, qual=10, ind=1)
```

```python
#Rests index for use in loops
career_batting.reset_index(inplace=True)
seasonal_batting.reset_index(inplace=True)
```

```python
#Data Engineering - Creating a total seasons column
tot_seasons = []
for i in range(0, len(career_batting)):
    
    count = 0
    for j in range(0, len(seasonal_batting)):
        if seasonal_batting['Name'][j] == career_batting['Name'][i]:
            count += 1
    tot_seasons.append(count)
            
```

```python
career_batting['Seasons'] = tot_seasons
```

### Getting Seasonal by Season Data for Batters with 10+ Seasons

```python
#Saving the names of every player who played at least 10 seasons
count = 0
names_for_forecasting = []
for i in range(0, len(career_batting)):
    if career_batting['Seasons'][i] > 9:
        names_for_forecasting.append(career_batting['Name'][i])
        count += 1
```

```python
#To give an idea of the columns involved
x = seasonal_batting.columns
for i in range(0, len(x)):
    print(x[i])
```

### Now interested in creating dataframes(For Time Series) of every player who played at least 10 seasons

```python
#Creating a list that holds a dataframe of season by season data, for every player in our names_for_forecasting list

#List to hold our dataframes
forecasting_dfs = []    

#Looks for a row with a matching player to names_for_forecasting, and adds certain metrics to a new row
#creating a dataframe of seasons for each player at a time
for i in range(0, len(names_for_forecasting)):
    
    df = pd.DataFrame()
    curr_name = names_for_forecasting[i]
    
    for j in range(0, len(seasonal_batting)):
        if seasonal_batting['Name'][j] == curr_name:
            curr_row = {'Date': seasonal_batting['Season'][j], 'Name': seasonal_batting['Name'][j], 'Age': seasonal_batting['Age'][j], 'AB': seasonal_batting['AB'][j], 'H': seasonal_batting['H'][j], 'G': seasonal_batting['G'][j], 'WAR': seasonal_batting['WAR'][j], 'RBI': seasonal_batting['RBI'][j], 'SO': seasonal_batting['SO'][j], 'BB': seasonal_batting['BB'][j], 'BABIP': seasonal_batting['BABIP'][j], 'GB%': seasonal_batting['GB%'][j], 'FB%': seasonal_batting['FB%'][j]} 
            df = df.append(curr_row, ignore_index=True)
    forecasting_dfs.append(df)
```

```python
#Veryifying code worked as intended, it did
forecasting_dfs[1]
```

```python
#Want to filter our list to only a batter who did hit last season

bat_2022 = []
for i in range(0, len(forecasting_dfs)):
    
    if forecasting_dfs[i]['Date'].max() == 2022:
        bat_2022.append(forecasting_dfs[i])
    
```

```python
bat_2022[0]
```

```python
#For loop just to get the names of each batter
just_names = []
for i in range(0, len(bat_2022)):
    just_names.append(bat_2022[i]['Name'][0])
```

```python
#Now, want to get all usable dataframes into time series formatting
for i in range(0, len(bat_2022)):
    bat_2022[i].set_index('Date', inplace = True)
    bat_2022[i].sort_index(inplace=True)
    
```

```python
#Making Stat Specific Dataframes
#Career Data - Key is to only include rate statisics, since not all pitches have the same workload in their respective roles
df_career = pd.DataFrame()
df_career['Seasons'] = career_batting['Seasons']
df_career['Age'] = career_batting['Age']
df_career['AVG'] = career_batting['AVG']
df_career['WAR'] = career_batting['WAR']
df_career['OBP'] = career_batting['OBP']
df_career['wOBA'] = career_batting['wOBA']
df_career['BABIP'] = career_batting['BABIP']
df_career['HR'] = career_batting['HR']
df_career['RBI'] = career_batting['RBI']
df_career['SLG'] = career_batting['SLG']
df_career['Contact%'] = career_batting['Contact%']
df_career['OPS'] = career_batting['OPS']
```
