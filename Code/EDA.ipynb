### EDA Of Pitching Data

```python
#Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading in Data
season = pd.read_csv('./season.csv')
career = pd.read_csv('./career.csv')
```

```python
#Starting with Seasonal Data
#Interested in correlations of our variables
season.corr()
```

Notable correlations:

Earned run average and war are strongly negatively correlated.
K/9 and War have a positive correlation.

```python
#Now Career Correlations
career.corr()
```

Notable correlations:
    
Now games played is correlated with war and saves, which intuitively makes sense.
War highly correlated with complete games as well.

```python
#Looking at some distributions
#Seasonal First
plt.hist(season['g'])
plt.title('Games Played in Season')
plt.ylabel('Freq of Players')
plt.xlabel('Games Played')
```

Approximately Normal, some possible outliers near 10.

```python
#Now Career
plt.hist(career['g'])
plt.title('Games Played in Career')
plt.ylabel('Freq of Players')
plt.xlabel('Games Played')
```

Very skewed right.

```python
#Seasonal War
plt.hist(season['war'])
plt.title('War Seasonal Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('War')
```

Approx Normal.

```python
#Career War
plt.hist(career['war'])
plt.title('War Career Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('War')
```

Very skewed right.

```python
#Seasonal ERA
plt.hist(season['era'])
plt.title('Era Seasonal Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('Era')
```

Approx normal.

```python
#Career ERA
plt.hist(career['era'])
plt.title('Era Career Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('Era')
```

A bit skewed right, but mostly normally distributed.

```python
#Seasonal k/9
plt.hist(season['k/9'])
plt.title('k/9 Sesonal Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('k/9')
```

Approx normal.

```python
#Career k/9
plt.hist(career['k/9'])
plt.title('k/9 Career Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('k/9')
```

Approx Normal.

```python
#Seasonal sv
plt.hist(season['sv'])
plt.title('Sv Sesonal Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('Sv')
```

A rare statistic for a player to have, possible to remove from data and replace with something else.

```python
#Career Sv
plt.hist(career['sv'])
plt.title('sv Sesonal Distribution')
plt.ylabel('Freq of Players')
plt.xlabel('sv')
```

Very skewed right.

```python
#Now Interested in Seasonal describe
season.describe()
```

Some Interesting Takeaways:

A player pictched on average about 32 games in a season.

Some Interesting Takeaways:
